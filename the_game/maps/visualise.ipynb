```python
import matplotlib.pyplot as plt
import networkx as nx
from new_map import build_graph

# Build the directed graph using existing logic
G = build_graph()

# Extract positions for drawing
pos = nx.get_node_attributes(G, "pos")

# Set up colors by node type
color_map = {
    1: "cornflowerblue",  # blue
    2: "red",             # events
    3: "green",           # branches
    4: "gold",            # gold/star
}
node_colors = [color_map.get(G.nodes[n].get("type", 0), "gray") for n in G.nodes]

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=1000, arrows=True, font_weight='bold')
plt.title("Game Board Graph Visualizer")
plt.axis("off")
plt.show()
```
